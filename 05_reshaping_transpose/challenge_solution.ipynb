Tasks
1. Convert into shape (4,6)
2. Convert into shape (6,4)
3. Print total elements
4. Print dimensions
5. Flatten the array
6. Transpose the (4,6) array
7. Print transpose shape
8. Create a (2,3,4) array
9. Flatten the (2,3,4) array
10. Verify total elements remain unchanged

❌ No loops allowed


```python
import numpy as np

X = np.arange(1,25)

#1
reshaped_4X6 = X.reshape(4, 6)
print("4X6 array:")
print(reshaped_4X6)

#2
reshaped_6X4 = X.reshape(6, 4)
print("\n6X4 array:")
print(reshaped_6X4)

#3
print("\nTotal elements:")
print(X.size)

#4
print("\nDimensions:")
print(reshaped_4X6.ndim)

#5 
flat = reshaped_4X6.flatten()
print("\nFlattened array:")
print(flat)

#6 
transposed = reshaped_4X6.T
print("\nTransposed array:")
print(transposed)

#7
print("\nTransposed shape:")
print(transposed.shape)

#8
reshaped_2X3X4 = X.reshape(2, 3, 4)
print("\n2X3X4 array:")
print(reshaped_2X3X4)

#9
flat_2X3X4 = reshaped_2X3X4.flatten()
print("\nFlattened 2X3X4 array:")
print(flat_2X3X4)

#10
print("\nOriginal shape:", reshaped_2X3X4.shape)
print("\nFlattened shape:", flat_2X3X4.shape)
print("\nVerification:", flat_2X3X4.size == reshaped_2X3X4.size)
```
