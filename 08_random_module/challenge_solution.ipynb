```python
import numpy as np

# Set Seed
np.random.seed(42)

# Random Integer Array
random_ints = np.random.randint(1, 101, size=10)

print("Random Integers:")
print(random_ints)

# Random Float Array
random_floats = np.random.rand(5)

print("\nRandom Floats:")
print(random_floats)

# Random Matrix
random_matrix = np.random.randint(1, 50, size=(3, 4))

print("\nRandom Matrix:")
print(random_matrix)

# Random Choice
choice = np.random.choice(['A', 'B', 'C', 'D'])

print("\nRandom Choice:")
print(choice)

# Mean
print("\nMean:")
print(np.mean(random_ints))

# Max
print("\nMaximum:")
print(np.max(random_ints))

# Min
print("\nMinimum:")
print(np.min(random_ints))
```
