```python
import numpy as np
x = np.array([
    [22, 25000, 1],
    [25, 30000, 2],
    [30, 50000, 5],
    [35, 70000, 8],
    [40, 90000, 12]
])

#1
print(x[x[:, 1] > 50000])

#2
print(x[x[:, 2] > 5])

#3
print(x[(x[:, 1] > 30000) & (x[:, 2] >= 5)])

#4
print(x[(x[:, 1] < 30000) | (x[:, 2] > 10)])

#5
print(x[~(x[:, 1] < 50000)])

#6
print(x[[0, 2, 4]])

#7
print(x[[0, 2, 4], 1])

#8 
print(x[[1, 3], 0])

#9
print(x[x[:, 0] > 30])

#10
print(x[(x[:, 0] >= 25) & (x[:, 0] <= 40)])


```
