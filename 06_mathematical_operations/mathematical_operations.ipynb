```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Array:")
print(arr)

print("\nSum:")
print(np.sum(arr))

print("\nMean:")
print(np.mean(arr))

print("\nMedian:")
print(np.median(arr))

print("\nMax:")
print(np.max(arr))

print("\nMin:")
print(np.min(arr))

print("\nStandard Deviation:")
print(np.std(arr))

print("\nVariance:")
print(np.var(arr))
```
