```python
import numpy as np

X = np.array([
    [22, 25000, 1],
    [25, 30000, 2],
    [30, 50000, 5],
    [35, 70000, 8],
    [40, 90000, 12]
])

print("Dataset:")
print(X)

# Sum
print("\nTotal Sum:")
print(np.sum(X))

# Column-wise Sum
print("\nColumn Sum:")
print(np.sum(X, axis=0))

# Row-wise Sum
print("\nRow Sum:")
print(np.sum(X, axis=1))

# Mean
print("\nColumn Mean:")
print(np.mean(X, axis=0))

# Max
print("\nColumn Maximum:")
print(np.max(X, axis=0))

# Min
print("\nColumn Minimum:")
print(np.min(X, axis=0))

# Standard Deviation
print("\nColumn Std:")
print(np.std(X, axis=0))

# Variance
print("\nColumn Variance:")
print(np.var(X, axis=0))
```

Tasks
1. Find total sum of all marks.
2. Find average marks.
3. Find maximum mark.
4. Find minimum mark.
5. Find median mark.
6. Find standard deviation.
7. Find variance.
8. Find subject-wise average (axis=0).
9. Find student-wise average (axis=1).
10. Find the highest average student.

❌ No loops allowed.

```python
import numpy as np

X = np.array([
    [85, 90, 88],
    [78, 82, 80],
    [92, 95, 91],
    [70, 75, 72]
])

#1
print("Sum of all elements in the dataset:")
print(np.sum(X))

#2
print("\nAverage Marks:")
print(np.mean(X))

#3
print("\nMaximum Marks:")
print(np.max(X))

#4
print("\nMinimum Marks:")
print(np.min(X))

#5
print("\nMedian Marks:")
print(np.median(X))

#6
print("\nStandard deviation of Marks:")
print(np.std(X))

#7
print("\nVariance of Marks:")
print(np.var(X))

#8
print("\nSubject-wise Average Marks (Axis=0):")
print(np.mean(X, axis=0))

#9
print("\nSubject-wise Average Marks (Axis=1):")
print(np.mean(X, axis=1))

#10
print("\nHighest Average Marks:")
print(np.max(np.mean(X, axis=0)))
```
