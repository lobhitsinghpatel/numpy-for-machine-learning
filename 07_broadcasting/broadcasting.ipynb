```python
import numpy as np
arr = np.array([10, 20, 30, 40])

print("Original array:")
print(arr)

print("\nAdd Scalar:")
print(arr + 5)

print("\nMultiply by Scalar:")
print(arr * 2)

# Example 2
X = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("\nDataset:")
print(X)

print("\nAdd 10:")
print(X + 10)

# Example 3
salary = np.array([25000, 30000, 50000])

bonus = np.array([2000])

print("\nSalary + bonus:")
print(salary + bonus)


```
