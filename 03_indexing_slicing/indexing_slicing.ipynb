```python
import numpy as np

X = np.array([
    [22, 25000, 1],
    [25, 30000, 2],
    [30, 50000, 5],
    [35, 70000, 8],
    [40, 90000, 12]
])

print("Dataset:")
print(X)

print("\nFirst row:")
print(X[0])

print("\nLast row:")
print(X[-1])

print("\nSalary column:")
print(X[:, 1])

print("\nExperience column:")
print(X[:, 2])

print("\nFirst 3 rows:")
print(X[0:3])

print("\nLast 2 rows:")
print(X[-2:])

print("\nAge and Salary Columns:")
print(X[:, 0:2])
```
