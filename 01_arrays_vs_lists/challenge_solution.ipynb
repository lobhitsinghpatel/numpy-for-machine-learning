```python
import numpy as np
x = np.array([
    [22, 25000, 1],
    [25, 30000, 2],
    [30, 50000, 5],
    [35, 70000, 8],
    [40, 90000, 12]
])
print("Shape:", x.shape)
print("No of Dimensions:", x.ndim)
print("Data Type:", x.dtype)
x[:, 1] = x[:, 1] + 5000
x[:, 2] = x[:, 2] + 1
print("Updated Dataset:\n", x)  
print("Total memory consumed by the array:", x.nbytes, "bytes")
salary_per_experience = x[:, 1] / x[:, 2]
print("\nSalary per experience:")
print(salary_per_experience)
```
